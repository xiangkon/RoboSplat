```python
import numpy as np
import matplotlib.pyplot as plt

def square_spiral_increasing_radius(steps, initial_step_size=0.1, radius_increment=0.1):
    x = [0]
    y = [0]
    
    directions = [(1, 0), (0, 1), (-1, 0), (0, -1)]
    direction_index = 0
    
    current_x, current_y = 0, 0
    
    step_size = initial_step_size
    for i in range(steps):
        dx, dy = directions[direction_index]
        current_x += dx * step_size
        current_y += dy * step_size
        
        x.append(current_x)
        y.append(current_y)
        
        # 每两步改变一次方向，并增加步长
        if (i + 1) % 5 == 0:
            direction_index = (direction_index + 1) % 4
            step_size += radius_increment
    
    return np.array(x), np.array(y)

x, y = square_spiral_increasing_radius(steps=200, initial_step_size=0.1, radius_increment=0.05)

plt.figure(figsize=(8, 8))
plt.plot(x, y, '-o')
plt.title("Square Spiral with Increasing Radius")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
theta = np.linspace(0, 10 * np.pi, 1000)  # 从0到10π，生成1000个点
a = 0.02  # 螺距参数（控制螺旋线疏密程度）

# 极坐标方程转换为直角坐标系
r = a * theta  # 螺旋线的半径随角度线性增长
x = r * np.cos(theta)  # x坐标
y = r * np.sin(theta)  # y坐标

# 绘制螺旋线
plt.figure(figsize=(8, 8))
plt.plot(x, y)
plt.scatter([0], [0], color='red', zorder=5)  # 标记原点(0,0)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # 添加x轴辅助线
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # 添加y轴辅助线
plt.xlabel('x')
plt.ylabel('y')
plt.title('Spiral Line with Origin at (0,0)')
plt.axis('equal')  # 保证x轴和y轴的比例相同，螺旋线显示正确
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# 参数设置
theta = np.linspace(0, 4*np.pi, 500)  # 从0到10π，生成1000个点
a = 0.006  # 螺距参数（控制螺旋线疏密程度）

# 极坐标方程转换为直角坐标系
r = a * theta  # 螺旋线的半径随角度线性增长
x = r * np.cos(theta)  # x坐标
y = r * np.sin(theta)  # y坐标

# 绘制螺旋线
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x, y)

# 绘制半径为0.06的圆
circle = Circle((0, 0), 0.06, edgecolor='red', facecolor='none', linestyle='--', linewidth=2)
ax.add_patch(circle)

# 标记原点(0,0)
ax.scatter([0], [0], color='red', zorder=5)

# 添加坐标轴辅助线
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
ax.axvline(0, color='black', linewidth=0.5, linestyle='--')

# 设置图形属性
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Spiral Line with Origin at (0,0) and a Circle of Radius 0.06')
ax.axis('equal')  # 保证x轴和y轴的比例相同，螺旋线显示正确
ax.grid(True)

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def square_spiral(steps, step_size=0.1):
    # 初始化数组来存储方形螺旋线的坐标
    x = []
    y = []
    
    # 方向向量：右、上、左、下
    directions = [(1, 0), (0, 1), (-1, 0), (0, -1)]
    
    # 当前方向索引
    direction_index = 0
    
    # 当前位置
    current_x, current_y = 0, 0
    
    # 添加起始点
    x.append(current_x)
    y.append(current_y)
    
    # 绘制方形螺旋线
    for i in range(steps):
        # 获取当前方向
        dx, dy = directions[direction_index]
        
        # 移动一步
        current_x += dx * step_size
        current_y += dy * step_size
        
        # 添加新位置
        x.append(current_x)
        y.append(current_y)
        
        # 每两步改变一次方向
        if (i + 1) % 2 == 0:
            direction_index = (direction_index + 1) % 4
    
    return np.array(x), np.array(y)

# 生成方形螺旋线的坐标
x, y = square_spiral(steps=200, step_size=0.1)

# 绘制方形螺旋线
plt.figure(figsize=(8, 8))
plt.plot(x, y)
plt.title("Square Spiral")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')  # 保持比例
plt.grid(True)
plt.show()
```
